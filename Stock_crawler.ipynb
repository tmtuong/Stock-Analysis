```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
```

#Overview

```python
link_base = 'https://www.cophieu68.vn/companylist.php?currentPage=number&o=s&ud=a'
```

```python
link_list = []
for i in range(1,79):
  link_page = link_base.replace('number',str(i))
  link_list.append(link_page) 
```

```python
def remove_tags(a):
  while a.find('<') != -1 and a.find(">") != -1:
    upper = a.find('<')
    lower = a.find(">")
    i = a[upper:lower+1]
    a = a.replace(i,'')
  while a.find('\n') != -1:
    a = a.replace('\n','')
  while a.find('\r') != -1:
    a = a.replace('\r','')
  while a.find('\t') != -1:
    a = a.replace('\t','')
  while a.find('  ') != -1:
    a = a.replace('  ','')
  return a

def remove_(a):
  while a.find('(') != -1 and a.find(")") != -1:
    upper = a.find('(')
    lower = a.find(")")
    i = a[upper:lower+1]
    a = a.replace(i,'')
  return a

def convert_type(a):
  x = str(a).split(',')
  k = ''.join(x)
  if k.find('%') != -1:
    k = k.replace('%','')
    k = eval(k)/100
    return k
  else: return k 
```

```python
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/105.0.0.0 Safari/537.36'}
link = link_base.replace('number','1')
r = requests.get(link, headers)
soup = BeautifulSoup(r.text, 'html.parser')
tb_data = soup.find_all('table')
tr = tb_data[3].find_all("tr")
headers_lst = []
for row in tr[0]:
  headers = row
  for col in headers:
    cell = remove_tags(str(col)).strip().encode('iso-8859-1').decode('utf-8')
    if cell != '' and cell != 'tỷ':
      headers_lst.append(cell)
df = pd.DataFrame(columns = headers_lst)
df
```

```python
df = pd.DataFrame(columns = headers_lst)
for i in range(0,78):
  link = link_list[i]
  r = requests.get(link, headers)
  soup = BeautifulSoup(r.text, 'html.parser')
  tb_data = soup.find_all('table')
  tr = tb_data[3].find_all("tr")

  for row in tr[1:]:
    data_lst = []
    data = row.find_all('td')
    for col in data:
      cell = remove_(remove_tags(str(col)).replace('\n',''))
      data_lst.append(cell.strip())
    df.loc[len(df)] = data_lst
  print(f'Page:{i+1} - Crawl succeed!!!')
```

```python
df
```

```python
for i in df['Tên Công Ty']:
  x = i.split(' ')
  for k in range(0,len(x)):
    try:
      x[k] = x[k].encode('latin-1','replace').decode('utf-8')
    except UnicodeDecodeError: 
      loc = x[k].find('Ä')
      if loc != -1:
        x[k] ='Đ'+ x[k][loc+2:]
      continue
  df['Tên Công Ty'] = df['Tên Công Ty'].replace(i,' '.join(x))
```

```python
not_stock = [x for x in df['Mã CK'] if len(x) != 3 or x[0] =='^']
for i in not_stock:
  df.drop(df[df['Mã CK']==i].index,inplace = True)
df
```

```python
for i in df.columns[4:]:
  try:
    df[i] = df[i].apply(convert_type)
  except AttributeError: continue
df.info()
```

#Detail

```python
company_link = 'https://www.cophieu68.vn/snapshot.php?id=company'
company_lst = []
for i in df['Mã CK']:
  link = company_link.replace('company',i.lower())
  company_lst.append(link)
print(company_lst)
print(len(company_lst))
```

```python
def to_vn(a):
  x = a.split(' ')
  if len(x) == 1:
    return a
  else:
    for k in range(0,len(x)):
      try:
        x[k] = x[k].encode('latin-1','replace').decode('utf-8')
      except UnicodeDecodeError: 
        loc = x[k].find('Ä')
        if loc != -1:
          x[k] ='Đ'+ x[k][loc+2:]
        continue
    return ' '.join(x)
```

```python
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/105.0.0.0 Safari/537.36'}
df1 = pd.DataFrame()
for i in company_lst:
  r = requests.get(i, headers)
  soup = BeautifulSoup(r.text,'html.parser')
  tb = soup.find_all('table')
  
  t1 = tb[1].find_all('tr') 
  t2 = tb[2].find_all('tr')
  data_lst = []
  for x in t1:
    rows = x.find_all('td')
    for row in rows:
      data_lst.append(to_vn(remove_tags(str(row))))
  for x in t2:
    rows = x.find_all('td')
    for row in rows:
      data_lst.append(to_vn(remove_tags(str(row))))
  data = {'Mã CK': str(i[i.find('=')+1:]).upper()}
  for i in range(0,len(data_lst),2):
    data[data_lst[i]] = data_lst[i+1]
  df1 = df1.append(data,ignore_index=True)
```

```python
df1
```

```python
def cut(a):
  x = a.split(' ')
  return x[0]
def min_split(a):
  x = a.split('/')
  return x[0].strip()
def max_split(a):
  x = a.split('/')
  return x[1].strip()


df1['EPS (ngàn)'] = df1['EPS'].apply(cut)
df1['PE (lần)'] = df1['PE'].apply(cut)
df1['Vốn thị trường (Tỷ)'] = df1['Vốn thị trường'].apply(cut)
df1['Khối Lượng Niêm yết (Triệu)'] = df1['Khối Lượng Niêm yết'].apply(cut)
df1['Giá sổ sách (Ngàn)'] = df1['Giá sổ sách'].apply(cut)
df1['Thấp nhất trong ngày GD gần nhất'] = df1['TN/CN'].apply(min_split)
df1['Cao nhất trong ngày GD gần nhất'] = df1['TN/CN'].apply(max_split)
```

```python
drop_cols = ['EPS','PE','Vốn thị trường','Khối Lượng Niêm yết','Giá sổ sách','TN/CN']
new_df = df1.drop(columns = drop_cols, inplace = False)
new_df['KLTB 13 tuần'] = new_df['KLTB 13 tuần'].apply(convert_type)
```

```python
for i in new_df.columns[1:]:
  try:
    new_df[i] = new_df[i].apply(convert_type)
  except AttributeError: continue
```
